```python
# Importing libraries
import pandas as pd
```

```python
# reading match data into a pandas dataframe
matches = pd.read_csv('matches.csv', index_col=0)
```

```python
matches['round'].value_counts()
```

```python
# cleaning the data for machine learning
# we need the predictors which aren't numeric to be used on numerical data for the prediction process

matches['date'] = pd.to_datetime(matches['date'])
```

```python
matches.dtypes
```

```python
# creating predictors for machine learning
# predictors to be used here include 'venue', 'opponent', 'time', 'day'
matches['venue_code'] = matches['venue'].astype('category').cat.codes
matches['opp_code'] = matches['opponent'].astype('category').cat.codes
matches['hour'] = matches['time'].str.replace(':.+', '', regex=True).astype('int')
matches['day_code'] = matches['date'].dt.dayofweek
```

```python
# setting up a target to be predicted e.g whether the team is going to win or not and for this, 
# the 'result' column will be used

matches['target'] = (matches['result']=='W').astype('int')
matches
```
